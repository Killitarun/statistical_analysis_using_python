## 3.Performing Inferential Statistics:

Inferential statistics is a branch of statistics that focuses on drawing conclusions about a population based on a sample of data. Instead of analyzing an entire population (which is often impractical or impossible), inferential statistics uses sampling to make predictions or inferences about the population.

```python
from scipy import stats
```

### Hypothesis Testing:

Hypothesis testing is used to determine whether there is enough evidence in a sample of data to infer that a certain condition holds for the entire population. The process involves:

```python
chosen_value = 20

# Perform a one-sample t-test
t_stat, p_value = stats.ttest_1samp(df['BMI'].dropna(), chosen_value)

print(f"T-Statistic: {t_stat}")
print(f"P-Value: {p_value}")
```

### Confidence intervals:

A confidence interval is a range of values, derived from the sample, that is likely to contain the population parameter with a certain degree of confidence (often 95%). For example, a 95% confidence interval for the mean cholesterol level might suggest that we are 95% confident that the true population mean lies within the interval.

```python
import numpy as np
BMI=df.BMI;
sample_mean = np.mean(BMI)
standard_error = stats.sem(BMI)
confidence_interval = stats.norm.interval(0.95, loc=sample_mean, scale=standard_error)

print(f"95% Confidence Interval for BMI: {confidence_interval}")
```

## Exploring Regression Analysis:

Regression analysis helps to understand relationships between variables. It allows us to predict the value of one variable (dependent variable) based on the values of one or more other variables (independent variables).

### Linear Regression

```python
import statsmodels.api as sm

# Define independent variable (add constant for intercept)
X = sm.add_constant(df['BMI'])

# Define dependent variable
y = df['Age']

# Fit linear regression model
model = sm.OLS(y, X).fit()

# Print model summary
print(model.summary())
```
